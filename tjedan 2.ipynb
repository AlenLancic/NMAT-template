# Lagrangeov interpolacijski polinom

Za točke ${T_i} = \left( {{x_i},{y_i}} \right),i = 0,1, \ldots ,n$ želimo polinom stupnja $n$ koji prolazi kroz svaku od njih, znamo da postoji jedinstveni takav ako su apscise točaka međusobno različite.

Za

$${\omega _j}\left( x \right) = \prod\limits_{j = 0,j \ne i}^n {\left( {x - {x_j}} \right)} $$

ste na predavanjima vidjeli kako se definira Lagrangeov interpolacijski polinom 

$${p_L}\left( x \right) = \sum\limits_{i = 0}^n {{y_i}\frac{{{\omega _i}\left( x \right)}}{{{\omega _i}\left( {{x_i}} \right)}}} $$

## Zadatak:
Izračunajte Lagrangeov interpolacijski polinom kroz točke ${T_0} = \left( {1,2} \right),{T_1} = \left( {3,4} \right),{T_2} = \left( {6,5} \right),{T_3} = \left( {7,8} \right)$.

## Rješenje:
Imamo

$${p_L}\left( x \right) = {y_0}\frac{{{\omega _0}\left( x \right)}}{{{\omega _0}\left( {{x_0}} \right)}} + {y_1}\frac{{{\omega _1}\left( x \right)}}{{{\omega _1}\left( {{x_1}} \right)}} + {y_2}\frac{{{\omega _2}\left( x \right)}}{{{\omega _2}\left( {{x_2}} \right)}} + {y_3}\frac{{{\omega _3}\left( x \right)}}{{{\omega _3}\left( {{x_3}} \right)}} = 2\frac{{{\omega _0}\left( x \right)}}{{{\omega _0}\left( 1 \right)}} + 4\frac{{{\omega _1}\left( x \right)}}{{{\omega _1}\left( 3 \right)}} + 5\frac{{{\omega _2}\left( x \right)}}{{{\omega _2}\left( 6 \right)}} + 8\frac{{{\omega _3}\left( x \right)}}{{{\omega _3}\left( 7 \right)}}$$

te računamo

$$\begin{gathered}
  {\omega _0}\left( x \right) = \left( {x - 3} \right)\left( {x - 6} \right)\left( {x - 7} \right) \Rightarrow {\omega _0}\left( 1 \right) = \left( {1 - 3} \right)\left( {1 - 6} \right)\left( {1 - 7} \right) =  - 60 \\
  {\omega _1}\left( x \right) = \left( {x - 1} \right)\left( {x - 6} \right)\left( {x - 7} \right) \Rightarrow {\omega _1}\left( 3 \right) = \left( {3 - 1} \right)\left( {3 - 6} \right)\left( {3 - 7} \right) = 24 \\
  {\omega _2}\left( x \right) = \left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 7} \right) \Rightarrow {\omega _2}\left( 6 \right) = \left( {6 - 1} \right)\left( {6 - 3} \right)\left( {6 - 7} \right) =  - 15 \\
  {\omega _3}\left( x \right) = \left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 6} \right) \Rightarrow {\omega _3}\left( 7 \right) = \left( {7 - 1} \right)\left( {7 - 3} \right)\left( {7 - 6} \right) = 24 \\ 
\end{gathered} $$

odakle je 

$$\begin{gathered}
  {p_L}\left( x \right) = \frac{2}{{ - 60}}\left( {x - 3} \right)\left( {x - 6} \right)\left( {x - 7} \right) + \frac{4}{{24}}\left( {x - 1} \right)\left( {x - 6} \right)\left( {x - 7} \right) + \frac{5}{{ - 15}}\left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 7} \right) + \frac{8}{{24}}\left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 6} \right) = \\
  \frac{{ - 1}}{{30}}\left( {x - 3} \right)\left( {x - 6} \right)\left( {x - 7} \right) + \frac{1}{6}\left( {x - 1} \right)\left( {x - 6} \right)\left( {x - 7} \right) + \frac{{ - 1}}{3}\left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 7} \right) + \frac{1}{3}\left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 6} \right) \\ 
\end{gathered} $$

# Newtonov interpolacijski polinom
Isto tako ste vidjeli da za 

$$p\left[ {{x_0}, \ldots ,{x_i}} \right] = \frac{{p\left[ {{x_1}, \ldots ,{x_i}} \right] - p\left[ {{x_0}, \ldots ,{x_{i - 1}}} \right]}}{{{x_i} - {x_0}}}$$ 

uz $p\left[ {{x_i}} \right] = p\left( {{x_i}} \right) = {y_i}$, se Newtonov interpolacijski polinom definira kao 

$${p_N}\left( x \right) = \sum\limits_{i = 0}^n {p\left[ {{x_0}, \ldots ,{x_i}} \right]\prod\limits_{j = 0}^{i - 1} {\left( {x - {x_j}} \right)} } $$

## Zadatak:
Izračunajte Newtonov interpolacijski polinom kroz točke ${T_0} = \left( {1,2} \right),{T_1} = \left( {3,4} \right),{T_2} = \left( {6,5} \right),{T_3} = \left( {7,8} \right)$
## Rješenje:
Zapišemo točke u tablicu na sljedeći način

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&{}&{}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{}&{}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{}&{}&{}&{} 
\end{array}$$

popunjavamo sljedeći prazni redak

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&{\frac{{4 - 2}}{{3 - 1}}}&{\frac{{5 - 4}}{{6 - 3}}}&{\frac{{8 - 5}}{{7 - 6}}}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{}&{}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{}&{}&{}&{} 
\end{array}$$

izračunamo vrijednosti

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&1&{\frac{1}{3}}&3&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{}&{}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{}&{}&{}&{} 
\end{array}$$

popunjavamo sljedeći prazni redak

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&1&{\frac{1}{3}}&3&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{\frac{{\frac{1}{3} - 1}}{{6 - 1}}}&{\frac{{3 - \frac{1}{3}}}{{7 - 3}}}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{}&{}&{}&{} 
\end{array}$$

izračunamo vrijednosti

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&1&{\frac{1}{3}}&3&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{\frac{{ - 2}}{{15}}}&{\frac{2}{3}}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{}&{}&{}&{} 
\end{array}$$

popunjavamo zadnji redak

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&1&{\frac{1}{3}}&3&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{\frac{{ - 2}}{{15}}}&{\frac{2}{3}}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{\frac{{\frac{2}{3} - \frac{{ - 2}}{{15}}}}{{7 - 1}}}&{}&{}&{} 
\end{array}$$

izračunamo vrijednost

$$\begin{array}{c}
  i&0&1&2&3 \\ 
\hline
  {{x_i}}&1&3&6&7 \\ 
  {p\left[ {{x_i}} \right]}&2&4&5&8 \\ 
  {p\left[ {{x_i},{x_{i + 1}}} \right]}&1&{\frac{1}{3}}&3&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}}} \right]}&{\frac{{ - 2}}{{15}}}&{\frac{2}{3}}&{}&{} \\ 
  {p\left[ {{x_i},{x_{i + 1}},{x_{i + 2}},{x_{i + 3}}} \right]}&{\frac{2}{{15}}}&{}&{}&{} 
\end{array}$$

Iz tablice, koristeći formulu, čitamo da je

$$\begin{gathered}
  {p_N}\left( x \right) = p\left[ {{x_0}} \right] + p\left[ {{x_0},{x_1}} \right]\left( {x - {x_0}} \right) + p\left[ {{x_0},{x_1},{x_2}} \right]\left( {x - {x_0}} \right)\left( {x - {x_1}} \right) + p\left[ {{x_0},{x_1},{x_2},{x_3}} \right]\left( {x - {x_0}} \right)\left( {x - {x_1}} \right)\left( {x - {x_2}} \right) = \\
  2 + 1 \cdot \left( {x - 1} \right) + \frac{{ - 2}}{{15}}\left( {x - 1} \right)\left( {x - 3} \right) + \frac{2}{{15}}\left( {x - 1} \right)\left( {x - 3} \right)\left( {x - 6} \right) \\ 
\end{gathered} $$

```python
def p_N(x):
    return 2 + 1*(x-1) + (-2/15)*(x-1)*(x-3) + (2/15)*(x-1)*(x-3)*(x-6)
```

```python
y_N = [p_N(x1[i]) for i in range(len(x1))]
```

```python
%matplotlib inline
plt.figure(figsize=(6, 6))
plt.plot(x1, y1, 'c.', label='${p_L}(x)$')
plt.plot(x1, y_N, 'b:', label='${p_N}(x)$')
plt.plot(x0, y0, 'ro', label='$T_i$')
plt.legend(fontsize = 16)
plt.show()
```
